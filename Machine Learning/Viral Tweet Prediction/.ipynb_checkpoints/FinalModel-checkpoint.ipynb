# Feature Transformation Train and Test

```python
def null_percentage(df):
    null_sum = df.isnull().sum()
    for i,name in enumerate(df.columns):
        if null_sum[i] > 0:
            print(f"The Percentage of Null Values in {name} is {(null_sum[i]/df.shape[0])*100} %")
```

```python
null_percentage(train_merged)
```

### Encoding The Categorical Columns

```python
######################################################## For Test topic id too 
from sklearn.preprocessing import LabelEncoder
categorical_features = ['tweet_topic_ids']

# make an encoder object
encoder = LabelEncoder()

# fit and transform feature x2
for col in categorical_features:
    encoder.fit(pd.concat([train_merged[col], test_merged[col]], axis=0, sort=False))
    train_merged[col] = encoder.transform(train_merged[col])
    test_merged[col] = encoder.transform(test_merged[col])
```

### Trying Out Other Models

```python
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC  
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import AdaBoostClassifier
# clf = SVC(kernel='linear') 
```

```python
lgb = LGBMClassifier()
cb = CatBoostClassifier(metric_period=100)
xgb = XGBClassifier()
dtc = DecisionTreeClassifier()
lr = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC(kernel='linear')
vc = VotingClassifier(estimators=10)
adc = AdaBoostClassifier()
```

```python
models = [lgb,cb,xgb,dtc,lr,knn,svm,vc,adc]
```

```python
list_of_model_performances = []
def fit_model(models):
    for model in models:
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        list_of_model_performances.append((model,accuracy_score(y_test,y_pred))) 
```

```python
fit_model(models)
```
